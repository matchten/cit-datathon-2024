```python
data_path = "./datasets/all_commodities.csv"

# Load your dataset
data = pd.read_csv(data_path)

# Filter for 'Sugar' in the 'Commodity' column
sugar_data = data[data['Commodity'] == 'Sugar']

# Convert the 'Date-Time' column to datetime format
sugar_data['Date-Time'] = pd.to_datetime(sugar_data['Date-Time'])

# Filter the data between 1990 and 2017
sugar_data = sugar_data[(sugar_data['Date-Time'] >= '1990-01-01') & (sugar_data['Date-Time'] <= '2017-12-31')]

# Group the data by year and calculate the average cost
sugar_data['Year'] = sugar_data['Date-Time'].dt.year
costs = sugar_data.groupby('Year')['Value'].mean()

# Reset the index to have 'Year' as a column
costs = costs.reset_index()

# Rename columns for clarity
costs.columns = ['Year', 'Average Cost']

diabetes = dd[(dd['Year'] >= 1990) & (dd['Year'] <= 2017)]

# Output the final DataFrame
print(costs)
print(diabetes)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the dataframes are named 'costs' for Average Costs and 'diabetes' for Diabetes Rates
# Ensure both dataframes have 'Year' as an integer for proper merging and comparison
costs['Year'] = costs['Year'].astype(int)
diabetes['Year'] = diabetes['Year'].astype(int)

# Merge the datasets on 'Year'
merged_data = pd.merge(costs, diabetes, on='Year')

# Creating the plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# First axis for Average Costs
color = 'tab:red'
ax1.set_xlabel('Year')
ax1.set_ylabel('Average Cost', color=color)
ax1.plot(merged_data['Year'], merged_data['Average Cost'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Second axis for Diabetes Percentage
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Diabetes Percentage', color=color)
ax2.plot(merged_data['Year'], merged_data['Percentage'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Third axis for Age-Adjusted Diabetes Percentage
ax3 = ax1.twinx()
color = 'tab:green'
ax3.spines['right'].set_position(('outward', 60))
ax3.set_ylabel('Age-Adjusted Diabetes Percentage', color=color)
ax3.plot(merged_data['Year'], merged_data['Age-Adjusted Percentage'], color=color)
ax3.tick_params(axis='y', labelcolor=color)

# Title and show
plt.title('Comparison of Sugar Prices, Diabetes Percentage, and Age-Adjusted Diabetes Percentage')
plt.savefig('./plots/sugar_prices_diabetes.png')
plt.show()
```

```python
import pandas as pd

# Assuming your merged DataFrame is named 'merged_data'
# Calculate the correlation matrix for the specific columns
correlation_matrix = merged_data[['Average Cost', 'Percentage']].corr()

# Extract the correlation coefficient between 'Average Cost' and 'Diabetes Percentage'
correlation_coefficient = correlation_matrix.at['Average Cost', 'Percentage']
print("Correlation Coefficient between Average Cost and Diabetes Percentage:", correlation_coefficient)
```

```python
import statsmodels.api as sm

data_path = "./datasets/diabetes_total_spending.csv"

# Load your dataset
spending = pd.read_csv(data_path)

# Define function to perform linear regression andgruously plot results
def analyze_trends(y, label):
    X = spending[['X.1']]
    X = sm.add_constant(X)  # adding a constant for the intercept
    model = sm.OLS(y, X).fit()
    print(f"Regression summary for {label}:\n{model.summary()}")
    
    # Plotting the trend
    plt.plot(spending['X.1'], y, label=label)
    plt.plot(spending['X.1'], model.predict(X), label=f'{label} Trend')

# Analyze trends for all groups
analyze_trends(spending['All enrollees'], 'All enrollees')
analyze_trends(spending['Diabetes with complications'], 'Diabetes with complications')
analyze_trends(spending['Diabetes without complication'], 'Diabetes without complication')

plt.xlabel('Year')
plt.ylabel('Average Cost')
plt.title('Average Annual Spending of Individuals with Health Coverage')
plt.legend()

plt.savefig('./plots/diabetes_cost.png')
plt.show()
```

```python
# Load the data
meat = pd.read_csv("./datasets/Meat_Stats_Meat_Production.csv")

# Check for missing values
print("Missing values per column:")
print(meat.isnull().sum())

# Option 1: Remove rows with any missing values
meat = meat.dropna()

# Save the cleaned data frame back to CSV if needed
meat.to_csv('path_to_cleaned_data.csv', index=False)

# Display the cleaned data (choose cleaned_df or filled_df depending on your choice above)
print(meat.head())

# Ensure 'Year' is an integer
meat['Year'] = meat['Year'].astype(int)

# Filter rows for the years 1980 to 2017
meat = meat[(meat['Year'] >= 1980) & (meat['Year'] <= 2017)]

# Display the first few rows of the filtered data to confirm
print(meat)

meat['Production'] = pd.to_numeric(meat['Production'].str.replace(',', '').str.replace(' ', ''), errors='coerce')

meat = meat.pivot_table(index=['Date', 'Year', 'Month'], columns='Animal', values='Production', aggfunc='sum')

# Reset the index to turn the multi-index into columns (if necessary)
meat.reset_index(inplace=True)

# Optionally, fill NaN values if any column might have missing values on some dates
meat.fillna(0, inplace=True)  # This fills missing entries with 0, adjust as necessary

# Display the first few rows to verify the transformation
print(meat.head())

# Convert columns that are not numerical but should be
columns_to_convert = ['Beef', 'Broilers', 'Lamb and Mutton', 'Other Chicken', 'Pork', 'Turkey', 'Veal']

# Group by 'Year' and calculate the mean for each type of meat
meat = meat.groupby('Year')[columns_to_convert].mean()

# Display the average annual production
print(meat)

diabetes = pd.read_csv("./datasets/diabetes_diagnosis.csv")

# Merge the dataframes on the 'Year' column
# Ensure both dataframes have 'Year' as an integer type for a successful merge
print(meat.index)
meat.reset_index(inplace=True)
print(meat. columns)
meat['Year'] = meat['Year'].astype(int)
diabetes['Year'] = diabetes['Year'].astype(int)

# Merge operation
df_merged = pd.merge(meat, diabetes, on='Year', how='left')

# Display the first few rows of the merged dataframe to verify the merge
print(df_merged.head())
```

```python
# Assuming df is your DataFrame loaded with the necessary data
correlation_matrix = df_merged.corr()

plt.figure(figsize=(10, 8))  # Set the figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Meat Production and Diabetes Percentage')
plt.show()
```
